```python
import csv
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

with open('nis_laser.csv') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    nisLaser = []
    for row in readCSV:
        for ele in row[:-1]:
            nisLaser.append(float(ele))

with open('nis_radar.csv') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    nisRadar = []
    for row in readCSV:
        for ele in row[:-1]:
            nisRadar.append(float(ele))
```

```python
laser95 = [5.991] * len(nisLaser)

figure(num=None, figsize=(10, 8), dpi=80, facecolor='w', edgecolor='k')

plt.title("Normalized Innovation Squared for LIDAR measuremnets")
plt.plot(nisLaser, label="NIS values")
plt.plot(laser95, label="95% line")
plt.ylabel("NIS laser")
plt.xlabel("k")
plt.legend()
plt.show()

count = len([i for i in nisLaser if i < 5.991]) 

percent = count/len(nisLaser)

print("Percentage of measurements below the 95% line -> ",  round(percent * 100,2),"%")
```

```python
radar95 = [7.815] * len(nisRadar)

figure(num=None, figsize=(10, 8), dpi=80, facecolor='w', edgecolor='k')

plt.title("Normalized Innovation Squared for RADAR measuremnets")
plt.plot(nisRadar, label="NIS values")
plt.plot(radar95, label="95% line")
plt.ylabel("NIS radar")
plt.xlabel("k")
plt.legend()
plt.show()

count = len([i for i in nisRadar if i < 7.815]) 

percent = count/len(nisLaser)

print("Percentage of measurements below the 95% line -> ",  round(percent * 100,2),"%")
```
